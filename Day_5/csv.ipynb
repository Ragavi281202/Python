used to return the csv file list format - reader()

```python
import csv
```

```python
with open("file.csv","r") as f:
    read=csv.reader(f)
    for i in read:
        print(i)
```

 Using DictReader:

 used to return the csv file in dictionary format
 
        csv.DictReader reads each row as an OrderedDict (for Python 3.6+) or a regular dictionary (for earlier versions).

```python
import csv
with open("file.csv","r") as f:
    read=csv.DictReader(f)
    for i in read:
        print(i)
```

write in csv

```python
data = [
    ['Name', 'Age', 'City'],
    ['Alice', 25, 'New York'],
    ['Bob', 30, 'Los Angeles'],
    ['Charlie', 22, 'Chicago'],
    ['vijay', 20, "kasi theatre"]
]
```

```python
with open("output.csv","w",newline="") as f:
    write=csv.writer(f)
    for i in data:
        write.writerow(i)
print("Done")
```

write in csv using dictwriter

```python
# Sample data as dictionaries
data = [
    {'Name': 'Alice', 'Age': 25, 'City': 'New York'},
    {'Name': 'Bob', 'Age': 30, 'City': 'Los Angeles'},
    {'Name': 'Charlie', 'Age': 22, 'City': 'Chicago'}
]

# Define CSV fieldnames (column names)
fieldnames = ['Name', 'Age', 'City']
with open("output1.csv","w",newline="") as f:
    write=csv.DictWriter(f,fieldnames=fieldnames)
    write.writeheader()
    for i in data:
        write.writerow(i)
print("Done")
```
